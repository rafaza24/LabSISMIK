```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def exact_solution(t):
    return np.exp(-2 * t)
```

```python
def euler_method(f, y0, t):
    y = np.zeros(len(t))
    y[0] = y0
    for i in range(1, len(t)):
        dt = t[i] - t[i-1]
        y[i] = y[i-1] + dt * f(t[i-1], y[i-1])
    return y
```

```python
def heun_method(f, y0, t):
    y = np.zeros(len(t))
    y[0] = y0
    for i in range(1, len(t)):
        dt = t[i] - t[i-1]
        y_pred = y[i-1] + dt * f(t[i-1], y[i-1])  # Euler prediction
        y[i] = y[i-1] + dt * (f(t[i-1], y[i-1]) + f(t[i], y_pred)) / 2  # Heun correction
    return y
```

```python
# ODE definition: y' = -2y
def f(t, y):
    return -2 * y
```

```python
# Parameters
y0 = 1
t0, t_end = 0, 5
n_points = 50
t = np.linspace(t0, t_end, n_points)
```

```python
# Solutions
y_exact = exact_solution(t)
y_euler = euler_method(f, y0, t)
y_heun = heun_method(f, y0, t)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, y_exact, label="Exact Solution", marker="*", color="black", linestyle="--")
plt.plot(t, y_euler, label="Euler Method", marker="o", linestyle="-")
plt.plot(t, y_heun, label="Heun Method",  linestyle="-")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Comparison of Numerical Methods")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definisi fungsi PDB
def dy_dt(t, y):
    return -2 * y

# Kondisi awal
y0 = [1]
t_span = (0, 5)  # Interval waktu
t_eval = np.linspace(t_span[0], t_span[1], 100)  # Titik evaluasi

# Menyelesaikan PDB
sol = solve_ivp(dy_dt, t_span, y0, t_eval=t_eval, method="euler")

# Plot solusi
plt.figure(figsize=(8, 5))
plt.plot(sol.t, sol.y[0], label=r"$\frac{dy}{dt} = -2y$", color="blue")
plt.title("Solusi PDB", fontsize=14)
plt.xlabel("t (waktu)", fontsize=12)
plt.ylabel("y (solusi)", fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```
